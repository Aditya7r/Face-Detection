### Import the OpenCV Package and read Image

#### Load the image return it in the form of a Numpy array. 

```python
import cv2

imagePath = "/Users/adityaradhakrishnan/Desktop/CodeClause/Emmanuel_Macron_2023_(cropped).jpg"

img = cv2.imread(imagePath) #read the image 

img.shape #dimensions
```

### Convert the Image to Grayscale
To improve computational efficiency, we first need to convert this image to grayscale before performing face detection on it

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # convert image to grayscale

gray_image.shape
```

### Load the Classifier

##### haarcascade_frontalface_default.xml - classifier is designed specifically for detecting frontal faces in visual input

```python
face_classifier = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)
```

### Perform the Face Detection

```python
face = face_classifier.detectMultiScale(
    gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40)
)

#detectMultiScale() method is used to identify faces
```

### Drawing a Bounding Box

```python
for (x, y, w, h) in face:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 4)
```

### Displaying the Image

##### To display the image with the detected faces, we first need to convert the image from the BGR format to RGB

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.imshow(img_rgb)
plt.axis('off')
```
